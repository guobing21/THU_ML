### 2 数据预处理

```python
def get_all_paper_id_to_contents(data):
    all_paper_id_to_contents = {}
    all_paper_ids = list(data.keys())
    for paper_id in all_paper_ids:
        paper_content = ""
        authors_content = data[paper_id]["authors"]
        for a in authors_content:
            paper_content += a["name"]
            paper_content += " "
            paper_content += a.get("org", "")
            paper_content += " "
        paper_content += data[paper_id]["title"]
        paper_content += " "
        paper_content += data[paper_id].get("abstract", "") if data[paper_id].get("abstract", "") else ""
        paper_content += " "
        keywords_content = data[paper_id].get("keywords", "")
        for k in keywords_content:
            paper_content += k
            paper_content += " "
        paper_content += data[paper_id]["venue"]
        paper_content += " "
        paper_content += str(data[paper_id].get("year", ""))
        all_paper_id_to_contents[paper_id] = paper_content
    return all_paper_id_to_contents
```

```python
all_paper_id_to_contents = get_all_paper_id_to_contents(data02)
```

```python
len(all_paper_id_to_contents)
```

```python
all_user_to_all_contents = {}
for user in data01.keys():
    temp = []
    for paper_id in data01[user]:
        temp.append(all_paper_id_to_contents[paper_id])
    if temp:
        all_user_to_all_contents[user] = temp
```

```python
len(all_user_to_all_contents)
```

### 3 转换为tfidf表征

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
for user in all_user_to_all_contents.keys():
    print(user)
    vectorized = TfidfVectorizer()
    X = all_user_to_all_contents[user]
    X = vectorized.fit_transform(X)
    all_user_to_all_contents[user] = X
```

### 4 采用DBSCAN聚类

```python
from sklearn.cluster import DBSCAN
```

```python
res = {}
for user in all_user_to_all_contents.keys():
    temp = []
    X = all_user_to_all_contents[user]
    y_pred = DBSCAN().fit(X)
    authors = set(y_pred.labels_)
    for a in authors:
        temp2 = []
        for i, y_label in enumerate(y_pred.labels_):
            if a == y_label:
                temp2.append(data01[user][i])
        temp.append(temp2)
    res[user] = temp
```

```python
res
```

### 5 保存结果

```python
import json
```

```python
with open("result.json", "w") as fw:
    json.dump(res, fw)
```
